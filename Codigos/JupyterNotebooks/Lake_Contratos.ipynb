```python
def CetusContratacoesGov(ANOMES_INFO):
    #ANOMES_INFO = 201301
    import urllib.request
    urllib.request.urlretrieve('http://www.portaltransparencia.gov.br/download-de-dados/compras/'+str(ANOMES_INFO), 'dados_contratos_' + str(ANOMES_INFO) + '.zip')

    import zipfile
    with zipfile.ZipFile('dados_contratos_' + str(ANOMES_INFO) + '.zip',"r") as zip_ref:
        zip_ref.extractall('dados_contratos_' + str(ANOMES_INFO))    
    
    nm_file1 = str(ANOMES_INFO)+'_Compras.csv'
    nm_file2 = str(ANOMES_INFO)+'_ItemCompra.csv'
    nm_file3 = str(ANOMES_INFO)+'_TermoAditivo.csv'

    #-----------------------  COMPRAS ------------------------------
    import pandas as pd
    import chardet
    with open('dados_contratos_' + str(ANOMES_INFO) +'/' + nm_file1, 'rb') as f:
        result = chardet.detect(f.read())  # or readline if the file is large
    df_compras = pd.read_csv('dados_contratos_' + str(ANOMES_INFO) +'/' + nm_file1, encoding=result['encoding'], error_bad_lines=False,sep=';', engine='python')
    df_compras['DT_INFO'] = ANOMES_INFO
    df_compras['PK_NMR_CTR'] = df_compras['Número do Contrato']     

    #-----------------------  ItemCompra -----------------------------
    with open('dados_contratos_' + str(ANOMES_INFO) +'/' + nm_file1, 'rb') as f:
        result = chardet.detect(f.read())  # or readline if the file is large
    df_ItemCompra = pd.read_csv('dados_contratos_' + str(ANOMES_INFO) +'/' + nm_file2, encoding=result['encoding'], error_bad_lines=False,sep=';', engine='python')
    df_ItemCompra['DT_INFO'] = ANOMES_INFO
    df_ItemCompra['PK_NMR_CTR'] = df_ItemCompra['Número Contrato']     

    #-----------------------  TermoAditivo -----------------------------
    with open('dados_contratos_' + str(ANOMES_INFO) +'/' + nm_file1, 'rb') as f:
        result = chardet.detect(f.read())  # or readline if the file is large
    df_TermoAditivo = pd.read_csv('dados_contratos_' + str(ANOMES_INFO) +'/' + nm_file3, encoding=result['encoding'], error_bad_lines=False,sep=';', engine='python')
    df_TermoAditivo['DT_INFO'] = ANOMES_INFO
    df_TermoAditivo['PK_NMR_CTR'] = df_TermoAditivo['Número Contrato']     
    
    return df_compras,df_ItemCompra,df_TermoAditivo
```
